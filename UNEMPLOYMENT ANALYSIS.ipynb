```python
# Strip leading/trailing spaces from the column names
df2.columns = df2.columns.str.strip()

# Plotting the bar chart (full dataset)
plt.figure(figsize=(20,10))
plt.bar(df2['Region'], df2['Estimated Unemployment Rate (%)'])

# Set xtick labels correctly with rotation
plt.xticks(rotation=90, ha='right')

# Adding labels and title
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by Region')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Strip leading/trailing spaces from the column names
df1.columns = df1.columns.str.strip()

# Check for NaN values in 'Region' and 'Estimated Unemployment Rate (%)'
print(df1[['Region', 'Estimated Unemployment Rate (%)']].isna().sum())

# Drop rows with NaN values in 'Region' or 'Estimated Unemployment Rate (%)'
df1 = df1.dropna(subset=['Region', 'Estimated Unemployment Rate (%)'])

# Ensure 'Region' column is of string type
df1['Region'] = df1['Region'].astype(str)

# Plotting the bar chart (full dataset)
plt.figure(figsize=(20,10))
plt.bar(df1['Region'], df1['Estimated Unemployment Rate (%)'])

# Set xtick labels correctly with rotation
plt.xticks(rotation=90, ha='right')

# Adding labels and title
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by time')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df2.columns = df2.columns.str.strip()

# Selecting only the numeric columns for correlation matrix
numeric_cols = df2.select_dtypes(include=[float, int])

# Calculating the correlation matrix
corr_matrix = numeric_cols.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis')

# Adding title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```

```python
# Remove leading space from 'Date' column name
df2.rename(columns={' Date': 'Date'}, inplace=True)

# Create the line plot
fig = px.line(df2, x='Date', y='Estimated Unemployment Rate (%)', 
              title='Unemployment Rate Over Time', 
              labels={'Estimated Unemployment Rate (%)': 'Unemployment Rate'},
              template='plotly_dark')

# Add the shape for the COVID-19 event period
covid_event = {'start_date': '2020-03-01', 'end_date': '2021-06-30', 'event': 'Peak of Pandemic'}
fig.add_shape(type='rect', 
              x0=pd.to_datetime(covid_event['start_date']), 
              x1=pd.to_datetime(covid_event['end_date']), 
              y0=0, y1=1, 
              fillcolor='red', opacity=0.3, layer='below', 
              line=dict(color='red', width=2))

# Update layout
fig.update_layout(xaxis_title='Date', yaxis_title='Unemployment Rate', legend_title='Region')
fig.update_xaxes(type='category')

# Show the plot
fig.show()
```

```python
fig = px.bar(df2, x='Region', y='Estimated Unemployment Rate (%)', color='Region',
             animation_frame='Date', title='Unemployment Rate Over Time',
             labels={'Estimated Unemployment Rate (%)': 'Unemployment Rate (%)'},
             template='plotly', color_discrete_sequence=px.colors.sequential.Viridis,
             category_orders={'Region': ['Region1', 'Region2', 'Region3']})

fig.update_layout(xaxis={'categoryorder': 'total descending'},
                  xaxis_title='Region', yaxis_title='Unemployment Rate (%)',
                  barmode='stack', showlegend=False)

fig.add_annotation(text='Impact of Unemployment Over Time',
                   xref='paper', yref='paper', x=0.5, y=1.05,
                   showarrow=False, font=dict(size=16, color='black'))

fig.show()
```
